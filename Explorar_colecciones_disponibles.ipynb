### OpenEO

La API estandarizada de openEO le permite acceder y procesar conjuntos de datos de observación de la Tierra en el ecosistema utilizando bibliotecas de programación intuitivas. El procesamiento se escala automáticamente en la infraestructura en nube que también aloja los datos. Esto permite una amplia gama de casos de uso, desde la investigación exploratoria hasta la producción a gran escala de mapas e información derivados de la observación de la Tierra. [OpenEO](https://dataspace.copernicus.eu/analyse/apis/openeo-api)
